### Data Cleaning (The rossmann drug store dataset!)

```python
# converting the date and promo columns to their appropraite data types
df_train.Date =  df_train["Date"].astype('datetime64[ns]')
df_train.Promo = df_train['Promo'].astype('bool')
```

```python
df_train.info()
```

```python
# Create a new column called Month to depict each month the store sale occurred
df_train['Month'] = pd.to_datetime(df_train['Date']).dt.month
# Calling the unique values
df_train['Month'].unique()
```

The month column has been created by extracting the particular month each store sale was recorded in the dataset to ensure our analysis is easier to come by, but the contents of the column are in integer format and should be converted to a text format.

```python
#Import the calendar module
import calendar
# use the apply function embedded with the lambda to convert the month from number format to text
df_train['Month'] = df_train['Month'].apply(lambda x: calendar.month_abbr[x])
df_train['Month'].unique()
```

The column has been converted from integer to calendar abbreviations for easy comprehension

```python
#Convert the Dayoftheweek the column from integer to string
df_train['DayOfWeek']= df_train['Date'].dt.day_name()
#Find unique values to confirm the changes
df_train['DayOfWeek'].unique()
```

The DayOfWeek column was encoded by numbers and i used the datetime function to convert them to text for easier apprehension.

## Exploratory Data Analysis

### Research Question 1 (Does promotions have an impact on Store sales?)

```python
#Filter the dataframe to seperate stores with and without promo
df_promo = df_train[df_train['Promo'] == True]
df_nopromo = df_train[df_train['Promo'] == False]
```

```python
df_promo.Sales.describe()
```

The number of store sales that occured with promo has 388080 entries, the average sales made per store with promo is 7991.15.

```python
df_nopromo.Sales.describe()
```

The number of store sales that occured without promo has 629129 entries,almost twice the entries of store sales that ocurred with promotions,the average sales made per store without promotion is 4406.05 which is significantly lesser than the store sales with promotions.

```python
#Create a uniform color for our vizualizations
base_color = sns.color_palette()[0]
```

```python
#use seaborn countplot to count the entries
sns.countplot(data= df_train,x ='Promo',color =base_color).set(title = 'Number of Store entries with and without promotions');
```

According to the visualization above the store entries without promotions are a little above 600,000 while store entries with promotions are a little under 400,000.

```python
#Creating a barplot
g = sns.barplot(data = df_train,y='Sales',x= 'Promo',color =base_color).set(title='AVERAGE STORE SALES PER PROMO STATUS');
```

According to the visualization above it depicts that the average store sales made by stores without promotions is a little above 4,000 per entry, the average sales is almost twice that at 8,000 per entry. Proving that promotions increases average store sales.

### Research Question 2  (Does promotions have an impact on Store Customers?)

```python
#Create a bar plot
g = sns.barplot(data = df_train,y='Customers',x= 'Promo',color = base_color)
g.set_title('Average Customers from Stores without and with Promotions ');
```

The visualization above depicts that stores without promotions have an average of 500 customers per entry, whilst the stores with promotions have an average of 800 customers per entry. Proving that promotions impact positively the number of customers in the stores.

### Research Question 3  (Which days of the week has the most sales on average?)

```python
# Create a list called cats to store the correct order of the day
cats = [ 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
# import the categoricaldtype function from the pandas.api.types module
from pandas.api.types import CategoricalDtype
cat_type = CategoricalDtype(categories=cats, ordered=True)
#Use astype function to change the data type to an ordered category
df_train['DayOfWeek'] = df_train['DayOfWeek'].astype(cat_type)
```

I ensured I ordered the DayOfWeek column to ensure my visualization doesn't give me issues while visualizing and they are ordered chronologically.

```python
#Create barplot
sns.barplot(x='Sales',y='DayOfWeek',data=df_train,color=base_color).set(title='Average Sales per day of the week');
```

According to the horizontal barplot above Monday generates the most sales with almost 8000 per entry on average and Friday generates which is a little shy of 7000 on an average in sales per store.

### Research Question 4  (Which days of the week has the most customers on average?)

```python
#Use seaborn to create barplot
g = sns.barplot(x='Customers',y='DayOfWeek',data = df_train,color = base_color)
g.set_title('Average customers per day of the week');
```

We can see a similar trend here where the day of the week that had the most customers Monday at an average of 800 customers per entry,Tuesdays and Fridays are a close second. Sunday has the least activity with less than 50 customers per entry.

### Research Question 5  (What is the trend of customers across a full year?)

```python
#Creating a list called cats than contains the abbrevations of all months in the correct order
cats = [ 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul','Aug','Sep','Oct','Nov','Dec']
#import the categoricaldtype function
from pandas.api.types import CategoricalDtype
#create a variable cat_type
cat_type = CategoricalDtype(categories=cats, ordered=True)
#setting the month column   
df_train['Month'] = df_train['Month'].astype(cat_type)
```

The process above converts the month column into an chronological order so it will be easier to track the traffic of customers accross the different years.

```python
sns.set()
#plotting the lineplot of the different years along each other
sns.lineplot(data=df_train, x='Month', y='Customers').set(title ='Trend of average customers recorded across a full year')
```

The visualization above shows the trend of average customers recorded per entry in every month of the year. We can observe as influx of people sky rocket towards the end of the year.

### Research Question 6  (What is the trend of average revenue generation accross a full year?)

```python
sns.set()
#plotting the lineplot of the different years along each other
sns.lineplot(data=df_train, x='Month', y='Sales').set(title ='Trend of average customers recorded across a full year');
```

The visualization above filters the average amount of sales per entry accross the different years 2013,2014 and 2015. We can observe a similar trend accross the years as the sales drastically increase towards the end of the year.

### Research Question 7  (What relationship does the customer and sales have?)

```python
#plotting a regplot
g = sns.regplot(data = df_train, x = 'Customers',y='Sales',scatter_kws={'alpha':1/40},x_jitter = 0.3)
g.set_title('SALES VS CUSTOMERS');
```

```python
df_train['Customers'].corr(df_train['Sales'])
```

The visualization above shows a relationship between Customers and Sales,after checking the corrrelation co-efficient it brought about a correlation value of 0.89 asserted that the positive correlation is a strong one meaning they both have significant effects on each other i.e a change in one will cause a change in the other.
